```python
import pandas as pd

#baca file csv
df = pd.read_csv('/content/CreditCard-Fraud-Detection_K-Means.csv')

df.head(3)
df.info()
```

```python
df.head()
```

```python
df = df.rename(columns={'credit_card': 'Credit_Card', 'city': 'City',
                        'state': 'State','zipcode': 'Zipcode','credit_card_limit': 'Credit_Card_Limit'})
```

```python
df.head()
```

```python
X = df.drop(['City', 'State'], axis=1)

X.head()
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import KMeans

#membuat list yang berisi inertia
clusters = []
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(X)
  clusters.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# membuat plot inertia
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
# membuat objek KMeans
km5 = KMeans(n_clusters=5).fit(X)
```

```python
# menambahkan kolom label pada dataset
X['Labels'] = km5.labels_
```

```python
# membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(8,4))
sns.scatterplot(x=X['Zipcode'], y=X['Credit_Card_Limit'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.show()
```

```python
# membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(8,4))
sns.scatterplot(x=X['Credit_Card'], y=X['Credit_Card_Limit'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.show()
```
